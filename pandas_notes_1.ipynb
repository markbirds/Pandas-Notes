### Data Structures in Pandas
[Back to top](#top)

```python
import pandas as pd
import numpy as np

series1 = pd.Series(np.arange(3),index=list('abc'))
series2 = pd.Series({'a':0,'b':1,'c':2})
series3 = pd.Series(3,index=np.arange(3))

print('Series 1:\n',series1)
print('\nSeries 2:\n',series2)
print('\nSeries 3:\n',series3)
```

```python
print('Accessing Values in Series')
print("\nseries1['a']: ",series1['a'])
print("\nseries2[['a','b']]:\n",series2[['a','b']])
print("\nseries3[1:3]:\n",series3[1:3])
```

```python
data1 = [['Alex',10],['Bob',12],['Clarke',13]]
df1 = pd.DataFrame(data1,columns=['Name','Age'])

data2 = {'Name':['Tom', 'Jack', 'Steve', 'Ricky'],'Age':[28,34,29,42]}
df2 = pd.DataFrame(data2)

print('DataFrame from Lists:\n',df1)
print('\nDataFrame from Dictionary:\n',df2)
```

<a id='access'></a>
### DataFrame: Accessing Values,Adding and Deletion of Rows and Columns
[Back to top](#top)

```python
print('Adding and Deleting Column')
print('\nSample DataFrame:\n',df2)

df2['Random'] = pd.Series(np.random.randint(1,11,4))
print('\nAdded new column "Random":\n',df2)

df2.pop('Random')
print('\nDeleting Random Column using pop function:\n',df2)
```

```python
df2['Random'] = pd.Series(np.random.randint(1,11,4))

print('Adding and Deleting Rows:')
print('\nSample Data:\n',df2)

row = pd.DataFrame({'Name':'Mark','Age':20,'Random':np.random.randint(1,11,1)})
df = df2.append(row,ignore_index=True)
print('\nAdding rows using append:\n',df)

row = pd.DataFrame({'Name':'Pat','Age':19,'Random':np.random.randint(1,11,1)})
df = pd.concat([df,row],ignore_index=True)
print('\nAdding rows using concat:\n',df)

df2 = df.drop([0,1])
print('\nDeleting 1st and 2nd Row:\n',df2)
```

```python
print('Column Selection - Period and Square Bracket can be used to access a column')
print('\nSample DataFrame\n',df)

print('\nName Column:\n',df.Name)
print('\nName and Random Column:\n',df[['Name','Random']])
print('\nColumn Selection using iloc and colon:\n',df.iloc[:,1:3])
```

```python
print('Row Selection')
print('\nSample DataFrame\n',df)

print('\nSelecting last row:\n',df.iloc[-1])

print('\nSelecting Name and Age in 2nd to 5th row:\n',df.iloc[1:5][['Name','Age']])
```

### DataFrame Basic Functionality
[Back to top](#top)

```python
data = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack']),
   'Age':pd.Series([25,26,25,23,30,29,23]),
   'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8])}

dataframe = pd.DataFrame(data)

print('Sample DataFrame:\n',dataframe)
print('\nNumber of elements:\n',dataframe.size)

print('\nTop 3 rows:\n',dataframe.head(3))
print('\nBottom 3 rows:\n',dataframe.tail(3))

df = dataframe.set_index('Name')
print('\nSet Name column as index:\n',df)
```

### pipe(), apply() and applymap()
[Back to top](#top)

```python
df = pd.DataFrame(np.random.randint(1,101,(7,3)),columns=['Random1','Random2','Random3'])

print('Sample DataFrame:\n',df)

print('\npipe() takes the whole dataframe as argument: df.pipe(lambda x: print(x))\n')
df.pipe(lambda x: print(x))

print('\nprint(df.pipe(lambda x: x*2)\n')
print(df.pipe(lambda x: x*2))
```

```python
print('Sample DataFrame:\n',df)

print('\napply() takes each column(default axis=0) as argument: df.apply(lambda x: print(x))\n')
df.apply(lambda x: print(x))

print('\nprint(df.apply(lambda x: x*2 if x.name == \'Random2\' else x))\n')
print(df.apply(lambda x: x*2 if x.name == 'Random2' else x))

print('\nprint(df.apply(lambda x: x*3 if x.name == 0 else x,axis=1))\n')
print(df.apply(lambda x: x*3 if x.name == 0 else x,axis=1))
```

```python
print('Sample DataFrame:\n',df)

print('\napplymap() takes each values as argument: df.applymap(lambda x: \'pandas\' if x>50 else x)\n')
print(df.applymap(lambda x: 'pandas' if x>50 else x))
```
