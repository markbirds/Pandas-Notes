### DatetimeIndex and resample()
[Back to top](#top)

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('sampleData2.csv',parse_dates=['Date'],index_col='Date')

df.head()
```

```python
print('Data type of the index:',type(df.index))
```

```python
print('Sales Data in 2016')

df['2016']
```

```python
print('Revenue in October 2015')

df['2015-10'].Revenue.head(10)
```

```python
print('Revenue,Cost and Profit from 2015 to 2016')

df['2015':'2016'][['Revenue','Cost','Profit']].sum()
```

```python
print('Using resample(): Monthly Revenue,Cost and Profit')

df[['Revenue','Cost','Profit']].resample('A').sum()
```

```python
print('Monthly Sales from different Countries:')

df.pivot_table(index='Date',columns='Country',values='Revenue',aggfunc='sum').resample('M').sum()
```

### Period and PeriodIndex
[Back to top](#top)

```python
y = pd.Period('2016')
m = pd.Period('2016-03')
d = pd.Period('2016-03-29')

y
```

```python
m
```

```python
d
```

```python
print(y+1)
print(m+1)
print(d+1)
```

```python
print('Year start time:',y.start_time)
print('Year end time:',y.end_time)
print('\nMonth start time:',m.start_time)
print('Month end time:',m.end_time)
print('\nDay start time:',d.start_time)
print('Day end time:',d.end_time)
```

```python
q1= pd.Period('2017Q1')
print('Quarter 1 start time:',q1.start_time)
print('Quarter 1 end time:',q1.end_time)

q3= pd.Period('2017Q3')
print('\nQuarter 3 start time:',q3.start_time)
print('Quarter 3 end time:',q3.end_time)
```

```python
print('Difference between date_range and period_range')

date = pd.date_range('2011', '2017', freq='q')
date
```

```python
date = pd.period_range('2011', periods=10, freq='q')
date
```

### Shifting
[Back to top](#top)

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/py/master/pandas/20_shift_lag/fb.csv',parse_dates=['Date'],index_col='Date')
df = df.rename(columns={'Price':'Sales'})
print(df)
```

```python
print('Adding Previous Day Sales:\n')
df['Yesterday Sale'] = df.Sales.shift(1)

print(df)
```

```python
print('Adding Sales Growth:\n')
df['Growth'] = df['Yesterday Sale'] - df['Sales']

df['Sales Growth'] = (df['Yesterday Sale'] - df['Sales'])/df['Yesterday Sale']*100

print(df)
```
