```python
import pandas as pd
import numpy as np

#Create a Dictionary of series
d = {'Name':pd.Series(['Tom','James','Ricky','Vin','Steve','Smith','Jack',
   'Lee','David','Gasper','Betina','Andres']),
   'Age':pd.Series([25,26,25,23,30,29,23,34,40,30,51,46]),
   'Rating':pd.Series([4.23,3.24,3.98,2.56,3.20,4.6,3.8,3.78,2.98,4.80,4.10,3.65])
}

#Create a DataFrame
df = pd.DataFrame(d)
print(df)
```

```python
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])

print(df)
print(df.apply(np.mean,axis=0))
print(df.apply(np.mean,axis=1))
```

```python
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df)
print(df.max())
print(df.min())
print(df.max()-df.min())
print(df.pipe(lambda x: x.max() - x.min()))
print(df.apply(np.mean))
```

```python
def adder(ele1,ele2):
    return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df)
print(df.pipe(adder,df))
print(df.apply(np.mean))
print(df.mean())
print(np.mean(df))
```

```python
df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df)
print(df.pipe(lambda x:x*100))
print(df.applymap(lambda x:x*100))
print(df.apply(np.mean))
```

```python
ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings',
   'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
   'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
   'Year': [2014,2015,2014,2015,2014,2015,2016,2017,2016,2014,2015,2017],
   'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}
df = pd.DataFrame(ipl_data)
print(df)
print(df.groupby('Team').groups)
```

```python
grouped = df.groupby('Year')
print(grouped)

for name,group in grouped:
    print(name)
    print(group)
    
print(grouped.get_group(2014))
```

```python
print(grouped['Points'].agg(np.mean))
```

```python
print(grouped.agg(np.size))
```

```python
print(grouped.agg([np.mean,np.sum]))
```

```python
grouped = df.groupby('Team')
score = lambda x: (x - x.mean()) / x.std()*10
print(grouped.transform(score))
```

```python
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
print(left)
print(right)
print(pd.merge(left, right, on='id', how='inner'))
print(left.append(right).reindex())
```
